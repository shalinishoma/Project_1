#### 17. Build a dataset with every possible (UserID, ProductID) pair (cross join)

```python
# Get the user ids and product ids from the dataframes. 
dfm1=pd.DataFrame(users['UserID'].unique())
dfm2=pd.DataFrame(products['ProductID'].unique())

# Add a Key column to both the dataframes with a value 1. 
dfm1['key'] = 1
dfm2['key'] = 1

# Join the two dataframes above on the key column. 
dfm3=pd.merge(dfm1,dfm2,how='outer',on='key')
del dfm3['key']
dfm3.rename(columns={'0_x':'UserID','0_y':'ProductID'},inplace=True)
dfm3
```

#### 18. Determine how much quantity of each product was purchased by each user

```python
# Join the dataframe from the previous result and transactions on User Id and Product Id fields.  
dfm4=pd.merge(dfm3,transactions,how='left',on=['UserID','ProductID'])

# Replace Null values with zeroes. 
dfm4['Quantity'].fillna(0, inplace=True)

# display the dataframe. 
dfm5=dfm4[['UserID','ProductID','Quantity']]
dfm5
```

#### 19. For each user, get each possible pair of pair transactions (TransactionID1, TransacationID2)

```python
# Create a tran1 dataframe with transactions and a key value of 1
tran1=transactions.copy()
tran1['key']=1

# Join tran1 to itself on key to get the required result. 
dfm5=pd.merge(tran1,tran1,how='inner',on=['UserID','key'])
del dfm5['key']
dfm5
```

```python
# Get columns with missing information
missing_info = list(usertrand2.columns[usertrand2.isnull().any()])
missing_info
```

```python
# Get count of missing data for each column in the dataframe
for col in missing_info:
        num_missing = usertrand2[usertrand2[col].isnull() == True].shape[0] 
        print('number missing for column {}: {}'.format(col, num_missing))
```

```python
# Get percentage of missing data in the dataframe
for col in missing_info:
        percent_missing = usertrand2[usertrand2[col].isnull() == True].shape[0] / usertrand2.shape[0] 
        print('percent missing for column {}: {}'.format( col, percent_missing))
```
